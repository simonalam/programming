Now do it again, with slightly less direction:

8. Create a grid of 100 equally spaced points ranging from -6.5 to 6.5.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 50)                 # Step 2: grid of 50 points from 0 to 1
y = np.log(x)                             # Step 3: natural log values
z = np.exp(x)                             # Step 3: exponential values

plt.scatter(x, y, label="Natural Log")    # Step 4: scatter plot for log
plt.scatter(x, z, label="Exponential")    # Step 4: scatter plot for exp

plt.xlabel("x")                           # Step 6: x-axis label
plt.ylabel("value")                       # Step 6: y-axis label
plt.title("Natural Log and Exponential Functions")  # Step 6: title
plt.legend(loc="lower right")             # Step 7: legend in lower right

plt.show()                                # Step 5: display the plot
```

```python
# Steps 8–12: sine and cosine line plot
x = np.linspace(-6.5, 6.5, 100)          # Step 8: grid of 100 points from -6.5 to 6.5
y = np.sin(x)                            # Step 9: sine values
z = np.cos(x)                            # Step 9: cosine values

plt.plot(x, y, label="Sine")             # Steps 10–11: line plot for sine
plt.plot(x, z, label="Cosine")           # Steps 10–11: line plot for cosine

plt.xlabel("x")                          # Step 12: x-axis label
plt.ylabel("value")                      # Step 12: y-axis label
plt.title("Sine and Cosine Functions")   # Step 12: title
plt.legend(loc="lower left")             # Step 12: legend in lower left

plt.show()
```
